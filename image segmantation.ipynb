```python
def image_seg(image_path,num_clusters):
    """ image_path  : the path of the image you want 
        num_clusters: is the number the of colors you want at the image """
    img = imread(image_path)
    plt.imshow(img)
    plt.title('The Normal Image')
    plt.show()
    X_IMG = img.reshape(-1,img.shape[2])
    Kmeans_seg = KMeans(n_clusters=int(num_clusters))
    Kmeans_seg.fit(X_IMG)
    img_segm = Kmeans_seg.cluster_centers_[Kmeans_seg.labels_]
    img_segm = img_segm.reshape(img.shape)
    plt.imshow((img_segm*255).astype(np.uint8))
    plt.title(f'Using {num_clusters} Cluster')
    plt.show()
```
